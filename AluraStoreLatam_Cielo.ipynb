### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
precios_tienda1 = tienda["Precio"].tolist()
precios_tienda2 = tienda2["Precio"].tolist()
precios_tienda3 = tienda3["Precio"].tolist()
precios_tienda4 = tienda4["Precio"].tolist()

```

```python
def suma(tienda:list=[0])->float:
  calculo = sum(tienda)
  return calculo
```

```python
# Listas de precios por tienda
precios_tiendas = [
    precios_tienda1,
    precios_tienda2,
    precios_tienda3,
    precios_tienda4
]

# Ciclo para calcular totales y mostrarlos
for i, precios in enumerate(precios_tiendas, start=1):
    total = suma(precios)
    print(f'El total de ventas de la tienda {i} es {total}')
```

La tienda de mayor número de ventas fue la tienda 1, con $1,150,880,400.

 Mientras que la tienda con menor número de ventas fue la tienda 4, con $1,038,375,700.

# 2. Ventas por categoría

```python
##Guardo todos los data frame de tiendas en una lista

tiendas = [tienda, tienda2, tienda3, tienda4]
```

```python
#Primero me gustaría ver todas las categorías que hay en las tiendas, y saber si son las mismas, para ello.

for i, tienda in enumerate(tiendas, start=1):
    categorias = tienda["Categoría del Producto"].unique()
    print(f"Categorías en Tienda {i}: {categorias}")
```

```python
#Una vez que verifico que sí son las mismas categorías, busco saber qué categoría, de acuerdo con el total de ventas, es más popular en cada tienda
#Para ello agrupo los datos por categoría y sumo los precios de cada categoría.



for i, tienda in enumerate(tiendas, start=1):
    total_x_categoria = tienda.groupby("Categoría del Producto")["Precio"].sum()
    print(f"Total de ventas por categoría en Tienda {i}:\n{total_x_categoria}\n")

```

```python
for i, tienda in enumerate(tiendas, start=1):
    tienda["Total Venta"] = tienda["Precio"] * tienda["Cantidad de cuotas"]
    total_por_categoria = tienda.groupby("Categoría del Producto")["Total Venta"].sum()
    print(f"Total de ventas por categoría en Tienda {i}:\n{total_por_categoria}\n")
```

En las cuatro tiendas, la categoría más popular de acuerdo con su total de ventas fue Electrónicos.

# 3. Calificación promedio de la tienda


```python
for i, tienda in enumerate(tiendas, start=1):
    promedio_calificacion = tienda["Calificación"].mean()
    print(f"Promedio de calificación en Tienda {i}: {promedio_calificacion:.2f}")
```

La tienda con mejor calificación, fue la tienda #4.

# 4. Productos más y menos vendidos

```python
#Considerando la cantidad de artículos vendidos (cantidad de cuotas) por producto, se calcula el total de venta por producto.


for i, tienda in enumerate(tiendas, start=1):
    tienda["Total Venta"] = tienda["Precio"] * tienda["Cantidad de cuotas"]
    ventas_por_producto = tienda.groupby("Producto")["Total Venta"].sum()

    top_3 = ventas_por_producto.nlargest(3)
    bottom_3 = ventas_por_producto.nsmallest(3)

    print(f"\nTienda {i} - Top 3 productos más vendidos:\n{top_3}\n")
    print(f"Tienda {i} - Top 3 productos menos vendidos:\n{bottom_3}\n")
```

# 5. Envío promedio por tienda

```python
for i, tienda in enumerate(tiendas, start=1):
    costo_promedio_envio = tienda["Costo de envío"].mean()
    print(f"Tienda {i}, costo promedio de envío: ${costo_promedio_envio:.2f}")
```

**Gráficos de interés**

```python
import matplotlib.pyplot as plt


for i, tienda in enumerate(tiendas, start=1):
    tienda["Total Venta"] = tienda["Precio"] * tienda["Cantidad de cuotas"]
    total_por_categoria = tienda.groupby("Categoría del Producto")["Total Venta"].sum()

    total_por_categoria.plot(kind="bar")
    plt.title(f"Ventas por categoría - Tienda {i}")
    plt.xlabel("Categoría")
    plt.ylabel("Ventas Totales ($)")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

Productos más vendidos por tienda

```python
for i, tienda in enumerate(tiendas, start=1):
    tienda["Total Venta"] = tienda["Precio"] * tienda["Cantidad de cuotas"]
    ventas_por_producto = tienda.groupby("Producto")["Total Venta"].sum()
    top_3 = ventas_por_producto.nlargest(3)

    top_3.plot(kind="barh", color='green')
    plt.title(f"Top 3 productos más vendidos - Tienda {i}")
    plt.xlabel("Ventas Totales ($)")
    plt.tight_layout()
    plt.show()
```

Promedio de calificación por tienda

```python
promedios = [tienda["Calificación"].mean() for tienda in tiendas]
plt.pie(promedios, labels=["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
        autopct="%1.1f%%", startangle=140, colors=['gold', 'lightcoral', 'lightskyblue', 'mediumseagreen'])
plt.title("Proporción de calificación promedio por tienda")
plt.axis("equal")  # para que el círculo no quede ovalado
plt.show()
```

**Ingresos totales por tienda**

```python
import matplotlib.pyplot as plt

# guardar los totales de cada tienda
ingresos_totales = []

# Ciclo para calcular totales y mostrarlos
for i, precios in enumerate(precios_tiendas, start=1):
    total = suma(precios)
    ingresos_totales.append(total)
    print(f'El total de ventas de la tienda {i} es {total}')


plt.bar(
    [f'Tienda {i}' for i in range(1, len(precios_tiendas)+1)],
    ingresos_totales,
    color='mediumseagreen'
)
plt.title("Ingresos totales por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos Totales ($)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

```
